```
#Odev3
#Girilen sayının faktöriyelini hesaplayan bir fonksiyon 
def Faktoriyel(x):
  if x == 1:
    return 1
  else:
    return (x * Faktoriyel(x-1))
while True:
  userInput=input("Faktöriyel hesabı için lütfen sayı giriniz! Çıkış İçin X'e basın")
  if userInput=='x':
    break
  try:
    print(Faktoriyel(int(userInput)))
  except ValueError:
    print("Lütfen Sayı Giriniz")
```
